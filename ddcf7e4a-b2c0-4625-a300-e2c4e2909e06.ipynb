```python
def plotting(data, title, color='#5470C6'):
    plt.figure(figsize=(15, 5))
    plt.style.use('ggplot')
    
    sns.lineplot(data=data, color=color, linewidth=2)
    
    plt.title(title, fontsize=16)
    plt.xlabel('–î–∞—Ç–∞', fontsize=12)
    plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    
    plt.show()
```

## –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

```python
def new_features(df, max_lag=10, rolling_mean_size=24, add_date_features=True):

    if not isinstance(df.index, pd.DatetimeIndex):
        raise ValueError("–ò–Ω–¥–µ–∫—Å –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å DatetimeIndex")
    
    data_copy = df.copy()
    target_col = 'num_orders'
    
    # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    if add_date_features:
        # —Å–∏–Ω—É—Å –∏ –∫–æ—Å–∏–Ω—É—Å –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è —Ü–∏–∫–ª–∏—á–Ω–æ—Å—Ç–∏, 23 –∏ 0 –Ω–∞—Ö–æ–¥—è—Ç—Å—è —Ä—è–¥–æ–º, –∞ –Ω–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–∞–ª–µ–∫–∏
        data_copy['month'] = data_copy.index.month
        data_copy['month_sin'] = np.sin(2 * np.pi * data_copy.index.month / 12)
        data_copy['month_cos'] = np.cos(2 * np.pi * data_copy.index.month / 12)
        
        data_copy['day'] = data_copy.index.day
        data_copy['dayofweek'] = data_copy.index.dayofweek
        data_copy['dayofweek_sin'] = np.sin(2 * np.pi * data_copy.index.dayofweek / 7)
        data_copy['dayofweek_cos'] = np.cos(2 * np.pi * data_copy.index.dayofweek / 7)
        
        data_copy['hour'] = data_copy.index.hour
        data_copy['hour_sin'] = np.sin(2 * np.pi * data_copy.index.hour / 24)
        data_copy['hour_cos'] = np.cos(2 * np.pi * data_copy.index.hour / 24)
        
        data_copy['is_weekend'] = data_copy.index.dayofweek.isin([5, 6]).astype(int)
    
    lags = [f'lag_{i}' for i in range(1, max_lag + 1)]
    for i, lag in enumerate(range(1, max_lag + 1)):
        data_copy[f'lag_{lag}'] = data_copy[target_col].shift(lag)
    
    data_copy['rolling_mean'] = data_copy[target_col].shift().rolling(rolling_mean_size).mean()
    data_copy['rolling_std'] = data_copy[target_col].shift().rolling(rolling_mean_size).std()
    data_copy['rolling_median'] = data_copy[target_col].shift().rolling(rolling_mean_size).median()
    
    data_copy.dropna(inplace=True)
    return data_copy
```

```python
models = []
lin_reg = LinearRegression()
```

```python
print(f'RMSE LinearRegression –Ω–∞ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏: {rmse_cv_lin_reg:.4f}')
print(f'RMSE LinearRegression –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ: {rmse_train_lin_reg:.4f}')
print('-'*50)

# –°–æ—Ö—Ä–∞–Ω—è–µ–º –º–æ–¥–µ–ª—å –∏ –º–µ—Ç—Ä–∏–∫–∏ –≤ —Å–ø–∏—Å–æ–∫
models.append([lin_reg, rmse_cv_lin_reg, rmse_train_lin_reg])
```

### RandomForestRegressor

```python
rf_model = RandomForestRegressor(random_state=RANDOM_STATE)
```

```python
rf_params = {
    'n_estimators': range(50, 100, 10),
    'max_depth': range(5, 20, 5),
}
```

```python
rf_grid = RandomizedSearchCV(
    rf_model, 
    rf_params, 
    scoring='neg_root_mean_squared_error',
    n_jobs=-1, 
    n_iter=3, 
    cv=cv, 
    verbose=5, 
    random_state=RANDOM_STATE
)
```

```python
def results(target, pred, rmse):
    result = target.to_frame().reset_index()
    result['prediction'] = pd.Series(pred)
    result.set_index('datetime', inplace=True)
    
    plt.figure(figsize=(12, 6))
    result[[target.name, 'prediction']].plot(style=['-', '--'])
    plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏—Ö –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π')
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(['–§–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ', '–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ'])
    
    plt.annotate(f'RMSE = {rmse:.4f}', xy=(0.05, 0.95), xycoords='axes fraction', 
                fontsize=12, bbox=dict(boxstyle="round", fc="white", alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    return result

```
